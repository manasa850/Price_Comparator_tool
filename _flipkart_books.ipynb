```python
import requests
from bs4 import BeautifulSoup as bs
import time

def scrape_page(url):
    page = requests.get(url, headers={'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'})
    soup = bs(page.content, 'html.parser')
    containers = soup.findAll("div", class_="_4ddWXP")
    
    for index, container in enumerate(containers, start=1):
        book_title_element = container.find("a", class_="s1Q9rs")
        book_title = book_title_element.text.strip() if book_title_element else "N/A"
        
        original_price_element = container.find("div", class_="_30jeq3")
        original_price = original_price_element.text.strip() if original_price_element else "N/A"
        
        discount_price_element = container.find("div", class_="_3I9_wc")
        discount_price = discount_price_element.text.strip() if discount_price_element else "N/A"
        
        discount_percent_element = container.find("div", class_="_3Ay6Sb")
        discount_percent = discount_percent_element.text.strip() if discount_percent_element else "N/A"
        
        prod_ratings_container = container.find("span", {"class": "_1lRcqv"})
        product_ratings_element = prod_ratings_container.find("div", {"class": "_3LWZlK"}) if prod_ratings_container else None
        product_ratings = product_ratings_element.text.strip() if product_ratings_element else "N/A"
        
        reviews_element = container.find("span", class_="_2_R_DZ") 
        reviews = reviews_element.text.strip() if reviews_element else "N/A"
        
        print(f"Details for Container {index}:")
        print("Book Title:", book_title)
        print("Original Price:", original_price)
        print("Discount Price:", discount_price)
        print("Discount Percentage:", discount_percent)
        print("Product Ratings:", product_ratings)
        print("Reviews:", reviews)
        print("------------------------------------------------------------------------------------------------------------------")
        
        # Add a delay of 2 seconds between requests
        time.sleep(1)

base_url = "https://www.flipkart.com/search?q=psychology+books&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&p%5B%5D=facets.price_range.from%3DMin&p%5B%5D=facets.price_range.to%3D1000"
page_number = 1
end_page = 3

for i in range(page_number, end_page + 1):
    url = f"{base_url}&page={i}"
    print(f"Scraping page {i}")
    scrape_page(url)
```

```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
import time

# Function to scrape page and store details in a list
def scrape_page(url, data):
    page = requests.get(url, headers={'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'})
    soup = bs(page.content, 'html.parser')
    containers = soup.findAll("div", class_="_4ddWXP")
    
    for container in containers:
        book_title_element = container.find("a", class_="s1Q9rs")
        Product_Name = book_title_element.text.strip() if book_title_element else "N/A"
        
        original_price_element = container.find("div", class_="_30jeq3")
        original_price = original_price_element.text.strip() if original_price_element else "N/A"
        
        discount_price_element = container.find("div", class_="_3I9_wc")
        discount_price = discount_price_element.text.strip() if discount_price_element else "N/A"
        
        discount_percent_element = container.find("div", class_="_3Ay6Sb")
        discount_percent = discount_percent_element.text.strip() if discount_percent_element else "N/A"
        
        prod_ratings_container = container.find("span", {"class": "_1lRcqv"})
        product_ratings_element = prod_ratings_container.find("div", {"class": "_3LWZlK"}) if prod_ratings_container else None
        product_ratings = product_ratings_element.text.strip() if product_ratings_element else "N/A"
        
        reviews_element = container.find("span", class_="_2_R_DZ") 
        reviews = reviews_element.text.strip() if reviews_element else "N/A"
        
        
        data.append([Product_Name, original_price, discount_price, discount_percent, product_ratings, reviews])

# Main script to iterate through pages and save data to CSV
base_url = "https://www.flipkart.com/search?q=psychology+books&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&p%5B%5D=facets.price_range.from%3DMin&p%5B%5D=facets.price_range.to%3D1000"
page_number = 1
end_page = 45

# List to store scraped data
all_data = []

for i in range(page_number, end_page + 1):
    url = f"{base_url}&page={i}"
    print(f"Scraping page {i}")
    scrape_page(url, all_data)
    time.sleep(2)  # Add a delay between requests

# Create a DataFrame from the list
columns = ["Book Title", "Original Price", "Discount Price", "Discount Percentage", "Product Ratings", "Reviews"]
df = pd.DataFrame(all_data, columns=columns)

# Save DataFrame to CSV
df.to_csv("flipkart_bookss.csv", index=False)
```
