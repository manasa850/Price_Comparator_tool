```python
#converting the generated data into csv file
import urllib.request
import time
import pandas as pd
from bs4 import BeautifulSoup

def scrape_page(url):
    headers = {
        'User-Agent': 'Mozilla/5.0 (iPad; CPU OS 12_2 like Mac OS X) AppleWebKit/605.1.15 (KHTML, like Gecko) Mobile/15E148'
    }
    request = urllib.request.Request(url, headers=headers)
    page = urllib.request.urlopen(request)
    page_html = page.read()
    page.close()
    page_soup = BeautifulSoup(page_html, "html.parser")
    containers = page_soup.findAll("div", {"class": "s-result-item"})
    
    data = []
    
    for container in containers:
        product_name = container.find("span", {"class": "a-text-normal"})
        prod_name = product_name.text.strip() if product_name else "N/A"

        if prod_name == "N/A":
            continue

        price_container = container.find("span", {"class": "a-offscreen"})
        original = price_container.text.strip() if price_container else "N/A"

        rating_container = container.find("span", {"class": "a-icon-alt"})
        prod_rating = rating_container.text.strip() if rating_container else "N/A"

        reviews_container = container.find("span", {"class": "a-size-base s-underline-text"})
        reviews_rating = reviews_container.text.strip() if reviews_container else "N/A"

        data.append({
            "Product_Name": prod_name,
            "Original Price": original,
            "Ratings": prod_rating,
            "Number of Reviews": reviews_rating
        })

    return data

base_url = "https://www.amazon.in/s"
query_params = {
    "k": "psychology+books",
}
url = f"{base_url}?{'&'.join([f'{key}={value}' for key, value in query_params.items()])}"

page_number = 1
all_data = []

for i in range(78):
    url_with_page = f"{url}&page={page_number}"
    print(f"Scraping page {page_number}")
    data = scrape_page(url_with_page)
    all_data.extend(data)
    page_number += 1
    time.sleep(1)

# Convert the list of dictionaries to a pandas DataFrame
df = pd.DataFrame(all_data)

# Save DataFrame to a CSV file
df.to_csv('amazon_books_dataa.csv', index=False)
```
