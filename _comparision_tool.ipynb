```python
import pandas as pd
from rapidfuzz import process

# Assuming you have two CSV files named amazon_books_dataa.csv and flipkart_bookss.csv
amazon_df = pd.read_csv("amazon_books_dataa.csv")
flipkart_df = pd.read_csv("flipkart_bookss.csv")

# Function to calculate weighted score for a given product
def calculate_weighted_score(product, platform):
    # Adjust the weights based on your preferences
    weight_rating = 0.4
    weight_reviews = 0.3
    weight_price = 0.3

    # Extract relevant columns based on the platform
    if platform == 'Amazon':
        rating = float(product['Ratings'].replace(' out of 5 stars', ''))
        reviews = int(product['Number of Reviews'].replace(',', ''))
        price = int(product['Original Price'].replace('₹', '').replace(',', ''))
    elif platform == 'Flipkart':
        rating = float(product['Product Ratings'])
        reviews = int(product['Reviews'].strip('()').replace(',', ''))
        price = int(product['Original Price'].replace('₹', '').replace(',', ''))

    # Calculate the weighted score
    weighted_score = (weight_rating * rating) + (weight_reviews * reviews) + (weight_price * price)

    return weighted_score

# Function to find the best product for a given book title
def find_best_product(book_title):
    # Find all matches in Amazon and Flipkart based on title
    amazon_matches = amazon_df[amazon_df['Product_Name'].apply(lambda x: process.extractOne(book_title, [x])[0] == book_title)]
    flipkart_matches = flipkart_df[flipkart_df['Book Title'].apply(lambda x: process.extractOne(book_title, [x])[0] == book_title)]

    # Calculate the weighted score for each match in Amazon
    amazon_matches['Weighted_Score'] = amazon_matches.apply(lambda x: calculate_weighted_score(x, 'Amazon'), axis=1)

    # Calculate the weighted score for each match in Flipkart
    flipkart_matches['Weighted_Score'] = flipkart_matches.apply(lambda x: calculate_weighted_score(x, 'Flipkart'), axis=1)

    # Find the product with the highest weighted score in both Amazon and Flipkart
    best_amazon_product = amazon_matches.loc[amazon_matches['Weighted_Score'].idxmax()]
    best_flipkart_product = flipkart_matches.loc[flipkart_matches['Weighted_Score'].idxmax()]

    return best_amazon_product, best_flipkart_product

# Example: Provide a book title as input
user_book_title = "The Power Of Positive Thinking"

# Find the best product in both Amazon and Flipkart based on title
best_amazon_product, best_flipkart_product = find_best_product(user_book_title)

# Display the details of the best product in Amazon
print("Best product in Amazon based on title:")
print(best_amazon_product)

# Display the details of the best product in Flipkart
print("\nBest product in Flipkart based on title:")
print(best_flipkart_product)
```
